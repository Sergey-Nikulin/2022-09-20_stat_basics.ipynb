### Задание 1. Загрузка данных

Изучить представленный набор данных на основе [описания его столбцов](https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.names), загрузить его и оставить 8 столбцов для дальнейшего изучения: `surgery?`, `Age`, `rectal temperature`, `pulse`, `respiratory rate`, `temperature of extremities`, `pain`, `outcome`. 

```python
horse_data = pd.read_csv('horse_data.csv', header=None, na_values='?')
# horse_data.head()
```

```python
df_new = horse_data.filter(items = [0,1,3,4,5,6,10,22])
df_new.head()
```

```python
df_new.rename(columns={0: "surgery", 1: "age", 3: "rectal_temperature", 4: "pulse", 5: "respiratory_rate", 
                       6: "temperature_of_extremities", 10: "pain", 22: "ourcome"}, inplace=True)
df_new.head()
```

### Задание 2. Первичное изучение данных
Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.

```python
df_new.describe()
```

```python
# в столбце "age" выбросом является значение "9" в кол-ве 24 строк.

q1 = df_new['age'].quantile(0.25)
q3 = df_new['age'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df_new[~df_new['age'].between(lower_bound, upper_bound, inclusive=True)]
# remove_outliers.describe()
# remove_outliers.info()
remove_outliers.head() 
```

```python
len(df_new[df_new['age'] == 9])
```

```python
# в столбце "pulse" выбросом является значения больше "146" в кол-ве 5 строк.

q2 = df_new['pulse'].quantile(0.25)
q4 = df_new['pulse'].quantile(0.75)
iqr = q4 - q2
lower_bound = q2 - (1.5 * iqr) 
upper_bound = q4 + (1.5 * iqr)
remove_outliers2 = df_new[~df_new['pulse'].between(lower_bound, upper_bound, inclusive=True)]
# remove_outliers2.describe()
# remove_outliers2.info()
remove_outliers2.head()
```

```python
len(df_new[df_new['pulse'] > 146])
```

```python
# в столбце "respiratory rate" выбросом является значения больше "60" в кол-ве 17 строк.

q5 = df_new['respiratory_rate'].quantile(0.25)
q6 = df_new['respiratory_rate'].quantile(0.75)
iqr = q6 - q5
lower_bound = q5 - (1.5 * iqr) 
upper_bound = q6 + (1.5 * iqr)
remove_outliers3 = df_new[df_new['respiratory_rate'].between(lower_bound, upper_bound, inclusive=True)]
# remove_outliers3.describe()
# remove_outliers3.info()
```

```python
len(df_new[df_new['respiratory_rate'] > 60])
```

### Задание 3. Работа с пропусками

Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
# кол-во пропусков %

print((df_new.isna().mean() * 100).round(2))

print('--------------------------------------------------')

print(df_new.info())
```

```python
# df_new.head(30)
```

```python
# 1 шаг: Меняем 9 на 2 в столбце 'age' , т.к. это явная ошибка.
df_new.replace({'age' : { 9 : 2}},  inplace=True)
```

```python
# 2 шаг: Заполняем пропуски в 'rectal temperature' ср.температурой по группам (операция/возраст).
print(df_new.groupby(['surgery', 'age'])['rectal_temperature'].median())
df_new['rectal_temperature'] = df_new['rectal_temperature'].fillna(df_new.groupby(['surgery', 'age'])['rectal_temperature'].transform('median'))
```

```python
df_new.info()
df_new.describe()
# df_new.head(30)
```

```python
# 3 шаг: Заполняем пропуски в 'rectal temperature' ср.температурой по группам (операция/возраст).
print(df_new.groupby(['surgery', 'age'])['temperature_of_extremities'].median())
df_new['temperature_of_extremities'] = df_new['temperature_of_extremities'].fillna(df_new.groupby(['surgery', 'age'])['temperature_of_extremities'].transform('median'))
```

```python
df_new.info()
df_new.describe()
# df_new.head(30)
```

```python
# 4 шаг: Оставляем только те строки в которых как минимум 6 значений заполнено.
df_new.dropna(thresh=6, inplace=True)
df_new.info()
df_new.describe()
# df_new.head(30)
```

```python
# 5 шаг: Заполняем пропуски в 'surgery' и 'ourcome' модой.
df_new['surgery'] = df_new['surgery'].fillna(df_new['surgery'].mode()[0])
df_new['ourcome'] = df_new['ourcome'].fillna(df_new['ourcome'].mode()[0])
df_new.info()
```

```python
# 6 шаг: Заполняем пропуски в 'rectal temperature' ср.температурой по группам (операция/возраст).
# print(df_new.groupby(['surgery', 'ourcome'])['respiratory_rate'].mean())
df_new['respiratory_rate'] = df_new['respiratory_rate'].fillna(df_new.groupby(['surgery', 'ourcome'])['respiratory_rate'].transform('mean'))
# df_new.info()
```

```python
# 7 шаг: Заполняем пропуски в 'pulse'.
print(df_new.groupby(['rectal_temperature', 'temperature_of_extremities'])['pulse'].mean())
df_new['pulse'] = df_new['pulse'].fillna(df_new.groupby(['rectal_temperature', 'temperature_of_extremities'])['pulse'].transform('mean'))
df_new.info()
```

```python
# 8 шаг: Заполняем пропуски в 'pain'.
df_new['pain'] = df_new['pain'].fillna(df_new.groupby(['ourcome', 'temperature_of_extremities'])['pain'].transform('mean'))
df_new.info()
```
